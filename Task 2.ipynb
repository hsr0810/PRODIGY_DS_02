```python
# MISSING_VALUES

missing_values = x.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
# DUBLICATE_VALUES

duplicates = x.duplicated().sum()
print("Duplicate Entries: ", duplicates)
```

```python
import matplotlib.pyplot as plt

# Plotting histograms for numerical features
x.hist(bins=20, figsize=(14, 10), layout=(3, 3),color="orange")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count plot for Sex
plt.figure(figsize=(10, 5))
sns.countplot(data=x, x='Sex')
plt.title('Count of Passengers by Sex')
plt.show()

# Count plot for Pclass
plt.figure(figsize=(10, 5))
sns.countplot(data=x, x='Pclass')
plt.title('Count of Passengers by Pclass')
plt.show()

# Count plot for Embarked
plt.figure(figsize=(10, 5))
sns.countplot(data=x, x='Embarked')
plt.title('Count of Passengers by Embarked')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = x.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
